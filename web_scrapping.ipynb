```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
import os
import requests
```

```python
url = "https://www.flipkart.com/search?q=vivo&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
#Used to request data from the server. INcase this doesnot works in a single GO,
# open the URL manually as this gets blocked often in office laptop
r = requests.get(url)
r
```

```python
html_doc = r.text
html_doc
```

```python
# Creating beautiful soup element
soup = BeautifulSoup(html_doc)
```

```python
pretty_soup = soup.prettify()
```

```python
p_text = soup.get_text()
p_text
```

```python
container = soup.findAll("div", {'class':'_2kHMtA'})
```

```python
len(container)
```

```python
container[3]
```

```python
con = container[3]
product_name = []
product_rating = []
product_discount = []
```

```python
name = con.find("div", {'class':'_4rR01T'}).text
(name)
```

```python
rating = con.find("div", {'class':'_3LWZlK'})
print(rating)
```

```python
for con in container:
    name = con.find("div", {'class':'_4rR01T'}).text
    product_name.append(name)
    rating = con.find("div", {'class':'_3LWZlK'})
    if(rating == None):
        product_rating.append("NO ratings")
    else:
        product_rating.append(rating.text) 
    discount = con.find("div", {'class':'_3Ay6Sb'})
    if(discount == None):
        product_discount.append("NO discount")
    else:
        product_discount.append(discount.span.text)
    
```

```python
(product_discount)
```

```python
dataf = pd.DataFrame({'Product_Name': product_name, 'Product_Rating': product_rating,'Product_discount': product_discount})
```
