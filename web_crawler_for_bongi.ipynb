```python
for i in range(50):
    
    val=data.iloc[i]['FK Assortment']
    brand = data.iloc[i]['Brand Name']
    brand = brand.replace(" ","+")
    
    url1 = "https://www.flipkart.com/search?q="+brand+"+saree&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.brand%255B%255D%3D"
    
    name=brand.split("+")
    l = len(name)

    url2 = url1 + name[0]

   
    for j in range(1,l):
        url2 = url2 + "%2B" + name[j]
        
        
    try:
        r = requests.get(url2)
    except requests.exceptions.ConnectionError as e:
        r = "No response"
    if(r!="No response"):
        html_doc = r.text
        soup = BeautifulSoup(html_doc)    # Creating beautiful soup element
        if soup.find("div", {'class':'_2WkVRV'}):
            brand_name = soup.find("div", {'class':'_2WkVRV'}).text

            assort = soup.find("div", {'class':'W_R1IA'}).span.text
            assort = assort.split(" ")[5]
            data.iloc[i,2:3]=assort
        else:
            data.iloc[i,2:3] = 'NA'
    else: 
        data.iloc[i,2:3] = 'NA'
```
